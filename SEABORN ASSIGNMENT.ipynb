```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Set a nice style for the plots
sns.set_style("whitegrid")
```

```python
# 1. Create a scatter plot with synthetic data
np.random.seed(0)
num_points = 50
x_scatter = np.random.rand(num_points)
y_scatter = 2 * x_scatter + 0.5 + np.random.randn(num_points) * 0.2
scatter_df = pd.DataFrame({'X': x_scatter, 'Y': y_scatter})

plt.figure(figsize=(8, 6))
sns.scatterplot(x='X', y='Y', data=scatter_df)
plt.title('Scatter Plot of Synthetic Data')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.show()
```

```python
# 2. Visualize the distribution of a numerical variable with random numbers
np.random.seed(1)
random_numbers = np.random.normal(loc=0, scale=1, size=100)
random_df = pd.DataFrame({'Random Numbers': random_numbers})

plt.figure(figsize=(8, 6))
sns.histplot(random_df['Random Numbers'], kde=True)
plt.title('Distribution of Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# 3. Compare different categories based on numerical values
category = ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C']
value = [10, 15, 8, 12, 18, 9, 11, 16, 10]
category_df = pd.DataFrame({'Category': category, 'Value': value})

plt.figure(figsize=(8, 6))
sns.barplot(x='Category', y='Value', data=category_df)
plt.title('Comparison of Categories by Value')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
# 4. Visualize the distribution of a numerical variable across different categories
np.random.seed(2)
category_dist = ['X'] * 30 + ['Y'] * 40 + ['Z'] * 30
numerical_value = np.random.normal(loc=5, scale=2, size=100)
distribution_df = pd.DataFrame({'Category': category_dist, 'Numerical Value': numerical_value})

plt.figure(figsize=(8, 6))
sns.boxplot(x='Category', y='Numerical Value', data=distribution_df)
plt.title('Distribution of Numerical Value Across Categories')
plt.xlabel('Category')
plt.ylabel('Numerical Value')
plt.show()
```

```python
# 5. Visualize the correlation matrix of a synthetic dataset using a heatmap
np.random.seed(3)
num_features = 4
num_samples = 100
correlation_data = np.random.rand(num_samples, num_features)
correlation_df = pd.DataFrame(correlation_data, columns=[f'Feature_{i+1}' for i in range(num_features)])
correlation_matrix = correlation_df.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Features')
plt.show()
```
